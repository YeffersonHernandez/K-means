```python
# Importar libreras necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
import os
import warnings

# Configuracin para evitar mltiples advertencias
warnings.filterwarnings('ignore')

# Paso 1: Anlisis exploratorio de los datos
# Carga del dataset
data = pd.read_csv('Mall_Customers.csv')

# Mostrar las primeras filas del dataset
print(data.head())

# Descripcin de las columnas del dataset para observar tendencias y distribuciones
print(data.describe())

# Identificacin de valores nulos y tipos de datos
print(data.info())

# Visualizacin de la distribucin de las variables relevantes
sns.pairplot(data)
plt.show()

# Distribucin de las variables individuales: Edad, Ingreso Anual, Puntaje de Gasto
plt.figure(figsize=(14, 5))
sns.histplot(data['Age'], kde=True, bins=20)
plt.title('Distribucin de Edad de los Clientes')
plt.show()

sns.histplot(data['Annual Income (k$)'], kde=True, bins=20)
plt.title('Distribucin de Ingreso Anual de los Clientes')
plt.show()

sns.histplot(data['Spending Score (1-100)'], kde=True, bins=20)
plt.title('Distribucin de Puntaje de Gasto de los Clientes')
plt.show()

# Paso 2: Preprocesamiento de los datos
# Eliminacin de columnas irrelevantes ('CustomerID')
data = data.drop(columns=['CustomerID'])

# Codificacin de la variable categrica 'Gender' en valores numricos
data['Gender'] = data['Gender'].map({'Male': 0, 'Female': 1})

# Identificacin de valores faltantes
print(data.isnull().sum())

# Anlisis y eliminacin de outliers utilizando el rango intercuartlico (IQR)
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
data = data[~((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))).any(axis=1)]

# Paso 3: Seleccin de Caractersticas
# Seleccin de caractersticas relevantes utilizando SelectKBest
X = data.drop(columns=['Spending Score (1-100)'])
y = data['Spending Score (1-100)']
selector = SelectKBest(score_func=f_classif, k=3)
X_new = selector.fit_transform(X, y)
selected_features = X.columns[selector.get_support()]
print(f"Caractersticas seleccionadas: {selected_features}")

# Escalado de las variables numricas
scaler = StandardScaler()
data_scaled = scaler.fit_transform(X_new)

# Paso 4: Reduccin de Dimensionalidad con PCA
# Aplicacin de PCA para reducir la dimensionalidad y facilitar la visualizacin
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data_scaled)

# Visualizacin de los resultados de PCA
plt.figure(figsize=(10, 6))
plt.scatter(data_pca[:, 0], data_pca[:, 1], alpha=0.5)
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.title('Visualizacin de Datos con PCA')
plt.show()

# Paso 5: Entrenamiento del modelo K-means
# Ajuste de la variable de entorno para evitar problemas de fuga de memoria en Windows con MKL
os.environ['OMP_NUM_THREADS'] = '1'

# Inicializacin del modelo con diferentes nmeros de clusters para identificar el ptimo
inertia = []
silhouette_scores = []
calinski_harabasz_scores = []
davies_bouldin_scores = []
range_clusters = range(2, 11)
for k in range_clusters:
    kmeans = KMeans(n_clusters=k, random_state=42, init='k-means++', n_init=10, max_iter=300)
    kmeans.fit(data_pca)
    inertia.append(kmeans.inertia_)
    labels = kmeans.labels_
    silhouette_scores.append(silhouette_score(data_pca, labels))
    calinski_harabasz_scores.append(calinski_harabasz_score(data_pca, labels))
    davies_bouldin_scores.append(davies_bouldin_score(data_pca, labels))

# Grfica del mtodo del codo para determinar el nmero ptimo de clusters
plt.figure(figsize=(10, 6))
plt.plot(range_clusters, inertia, marker='o')
plt.xlabel('Nmero de Clusters')
plt.ylabel('Inercia')
plt.title('Elbow Method para determinar el nmero ptimo de clusters')
plt.show()

# Grfica del coeficiente de Silhouette para evaluar la calidad del clustering
plt.figure(figsize=(10, 6))
plt.plot(range_clusters, silhouette_scores, marker='o', color='green')
plt.xlabel('Nmero de Clusters')
plt.ylabel('Coeficiente de Silhouette')
plt.title('Coeficiente de Silhouette para diferentes valores de k')
plt.show()

# Grfica del ndice de Calinski-Harabasz para evaluar la separacin de los clusters
plt.figure(figsize=(10, 6))
plt.plot(range_clusters, calinski_harabasz_scores, marker='o', color='blue')
plt.xlabel('Nmero de Clusters')
plt.ylabel('ndice de Calinski-Harabasz')
plt.title('ndice de Calinski-Harabasz para diferentes valores de k')
plt.show()

# Grfica del ndice de Davies-Bouldin para evaluar la calidad del clustering
plt.figure(figsize=(10, 6))
plt.plot(range_clusters, davies_bouldin_scores, marker='o', color='red')
plt.xlabel('Nmero de Clusters')
plt.ylabel('ndice de Davies-Bouldin')
plt.title('ndice de Davies-Bouldin para diferentes valores de k')
plt.show()

# Paso 6: Seleccin del nmero ptimo de clusters
# Seleccin del nmero ptimo de clusters basado en las grficas anteriores
optimal_k = 5

# Entrenamiento del modelo K-means con el nmero ptimo de clusters
kmeans_final = KMeans(n_clusters=optimal_k, random_state=42, init='k-means++', n_init=10, max_iter=300)
kmeans_final.fit(data_pca)

# Agregar las etiquetas de cluster al dataframe original
data['Cluster'] = kmeans_final.labels_

# Visualizacin de los clusters resultantes
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data_pca[:, 0], y=data_pca[:, 1], hue=data['Cluster'], palette='viridis', alpha=0.6)
plt.title('Clusters Resultantes del Modelo K-means')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.legend(title='Cluster')
plt.show()

# Paso 7: Interpretacin de los clusters
# Clculo de estadsticas descriptivas para cada cluster
cluster_summary = data.groupby('Cluster').mean()
print(cluster_summary)


# Resumen de mejoras aplicadas
print("\nSe realizaron las siguientes mejoras:")
print("1. Se ajustaron los hiperparmetros del modelo K-means: init='k-means++', n_init=10, max_iter=300.")
print("2. Se aadieron mtricas adicionales de evaluacin: ndice de Davies-Bouldin.")
print("3. Se realiz un anlisis de outliers utilizando el rango intercuartlico (IQR).")
print("4. Se graficaron mtricas adicionales como el ndice de Davies-Bouldin para una evaluacin ms completa del modelo.")
print("5. Se realiz una seleccin explcita de caractersticas relevantes usando SelectKBest.")
```
